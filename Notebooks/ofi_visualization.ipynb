Import Libraries and Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../Data/ofi_all_features.csv", parse_dates=['timestamp'])
```

Style and Symbol to Visualize

```python
sns.set(style='whitegrid')

symbol = 'AAPL'
df_plot = df[df['symbol'] == symbol]
```

Plot 1: Best-Level OFI

```python
plt.figure(figsize=(14, 4))
plt.plot(df_plot['timestamp'], df_plot['ofi_best'], label='Best-Level OFI', color='tab:blue')
plt.title(f"Best-Level OFI over Time for {symbol}")
plt.xlabel("Timestamp")
plt.ylabel("OFI Value")
plt.legend()
plt.tight_layout()
plt.show()
```

Plot 2: Multi-Level OFI (depth 09) over time

```python
plt.figure(figsize=(14, 6))
for i in range(10):
    plt.plot(df_plot['timestamp'], df_plot[f'ofi_lvl_{i}'], label=f'Level {i}', alpha=0.7)

plt.title(f"Multi-Level OFI (depth 09) over Time for {symbol}")
plt.xlabel("Timestamp")
plt.ylabel("OFI Value")
plt.legend(ncol=5)
plt.tight_layout()
plt.show()
```

Heat Map: Multi-Level OFI (depth 0-9)

```python
multi_ofi_matrix = df_plot.set_index('timestamp')[[f'ofi_lvl_{i}' for i in range(10)]].T
plt.figure(figsize=(14, 5))
sns.heatmap(multi_ofi_matrix, cmap="RdBu_r", center=0, cbar_kws={'label': 'OFI'})
plt.title(f"Multi-Level OFI Heatmap for {symbol}")
plt.ylabel("Order Book Depth Level")
plt.xlabel("Timestamp")
plt.tight_layout()
plt.show()
```

Plot 3: Integrated OFI (PCA)

```python
plt.figure(figsize=(14, 4))
plt.plot(df_plot['timestamp'], df_plot['ofi_integrated'], label='Integrated OFI', color='tab:green')
plt.title(f"Integrated OFI over Time for {symbol}")
plt.xlabel("Timestamp")
plt.ylabel("Integrated OFI")
plt.legend()
plt.tight_layout()
plt.show()
```

Plot 4: Cross-Asset OFI

```python
plt.figure(figsize=(14, 4))
plt.plot(df_plot['timestamp'], df_plot['ofi_cross_asset'], label='Cross-Asset OFI', color='tab:red')
plt.title(f"Cross-Asset OFI over Time for {symbol}")
plt.xlabel("Timestamp")
plt.ylabel("Cross-Asset OFI")
plt.legend()
plt.tight_layout()
plt.show()
```
